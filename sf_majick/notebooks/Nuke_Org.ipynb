```python
import requests

def delete_records_one_by_one(sf, object_api_name, record_ids, api_version="v59.0"):
    """
    Deletes Salesforce records individually using REST.
    """

    if not record_ids:
        print(f"No records to delete for {object_api_name}. Skipping.")
        return

    headers = sf.headers()

    for idx, record_id in enumerate(record_ids, start=1):
        url = f"{sf.instance_url}/services/data/{api_version}/sobjects/{object_api_name}/{record_id}"

        response = requests.delete(url, headers=headers)

        if response.status_code in (200, 204):
            print(f"✅ [{idx}/{len(record_ids)}] Deleted {object_api_name} {record_id}")
        else:
            print(
                f"❌ [{idx}/{len(record_ids)}] Failed deleting {object_api_name} {record_id}: "
                f"{response.status_code} {response.text}"
            )
```
